```python
import numpy as np
import pandas as pd
import h5py
import pylink
import hedfpy
from eyelinkio import read_edf
```

```python
# define filename
subject = "002"       # change to true subject number
session = "1"       # change
version = "2"       # change to true version number

subject2 = "01"
session2 = "2"
version2 = "5"

subject3 = "03"
session3 = "1"
version3 = "3"

filename = f'Logs/sub-{subject}/sub-{subject}_ses-{session}_v-{version}/sub-{subject}_ses-{session}_v-{version}.hdf5'
edf_filename = f'Logs/sub-{subject}/sub-{subject}_ses-{session}_v-{version}/sub-{subject}_ses-{session}_v-{version}.edf'

filename2 = f'Logs/sub-{subject2}/sub-{subject2}_ses-{session2}_v-{version2}/sub-{subject2}_ses-{session2}_v-{version2}.hdf5'
edf_filename2 = f'Logs/sub-{subject2}/sub-{subject2}_ses-{session2}_v-{version2}/sub-{subject2}_ses-{session2}_v-{version2}.edf'

filename3 = f'Logs/sub-{subject3}/sub-{subject3}_ses-{session3}_v-{version3}/sub-{subject3}_ses-{session3}_v-{version3}.hdf5'
edf_filename3 = f'Logs/sub-{subject3}/sub-{subject3}_ses-{session3}_v-{version3}/sub-{subject3}_ses-{session3}_v-{version3}.edf'

# check the contents of the file
print("Contents of file 1:")
with h5py.File(filename, 'r') as f:
    f.visititems(lambda name, obj: print(name, ':', obj))
print("\nContents of file 2:")
with h5py.File(filename2, 'r') as f:
    f.visititems(lambda name, obj: print(name, ':', obj))
print("\nContents of file 3:")
with h5py.File(filename3, 'r') as f:
    f.visititems(lambda name, obj: print(name, ':', obj))
```

```python
#create dataframes from the trials, trial_phases, parameters 
with h5py.File(filename, 'r') as f:
    trials = pd.DataFrame(f[f'sub-{subject}_ses-{session}_v-{version}.edf/trials'][:])
    trial_phases = pd.DataFrame(f[f'sub-{subject}_ses-{session}_v-{version}.edf/trial_phases'][:])
    trial_parameters = pd.DataFrame(f[f'sub-{subject}_ses-{session}_v-{version}.edf/parameters'][:])

# print the first few rows of each dataframe to check
print("Trials:")
print(trials.head(10))
print("\nTrial Phases:")
print(trial_phases.head(10))
print("\nTrial Parameters:")
print(trial_parameters.head(10))
```

```python
# check raw data
with h5py.File(filename3, 'r') as f:
    block1_items = pd.DataFrame(f[f'sub-{subject3}_ses-{session3}_v-{version3}.edf/block_0/block0_items'][:])
    block1_values = pd.DataFrame(f[f'sub-{subject3}_ses-{session3}_v-{version3}.edf/block_0/block0_values'][:])

print("\nBlock 1 Items:")
print(block1_items.head(10))
print("\nBlock 1 Values:")
print(block1_values.head(10))
```
